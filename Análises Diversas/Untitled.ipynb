```python
import pandas as pd      
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel ('gpt.xlsx', sheet_name='Dados GPT')
```

```python
df.head()
```

```python
df.set_index = ('COD')
```

```python
df['COD'].unique()
```

```python
df_cod_SETUP = df[(df['COD'] == 192) & (df['TEMPO (MIN)'] < 100) ]
```

```python
df_cod_SETUP.describe()
```

```python
df_cod_SETUP['TEMPO (MIN)'].median()
```

```python
plt.figure(figsize=(20,10)) 
plt.hist(df_cod_SETUP['TEMPO (MIN)'], bins=25)
plt.grid(axis='y', alpha=0.4)
plt.xlabel('Tempo (Min)',fontsize=15)
plt.ylabel('Frequência (Qtd)',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('Frequência (Qtd)',fontsize=15)
plt.title('Histograma dos Tempos Apontados de Setup',fontsize=25)
plt.show()
```
